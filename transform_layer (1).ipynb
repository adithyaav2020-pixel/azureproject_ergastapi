```python
# ADLS access permission
storage_account = "<storage_account_name>"
access_key = "<storage_account_key>"

spark.conf.set(
    f"fs.azure.account.key.{storage_account}.dfs.core.windows.net",
    access_key
)
```

```python
raw_path = "abfss://raw@driverf1storageprject.dfs.core.windows.net/drivers"

df = spark.read.format("delta").load(raw_path)

display(df)
```

```python
processed_df = df
```

```python
processed_df = df.dropDuplicates()
```

```python
processed_df.write \
  .format("delta") \
  .mode("overwrite") \
  .option("overwriteSchema", "true") \
  .save("abfss://processed@driverf1storageprject.dfs.core.windows.net/drivers")
```
