```python
presentation_path = "abfss://presentation@driverf1storageprject.dfs.core.windows.net/drivers"

df = spark.read.format("delta").load(presentation_path)

display(df)
```

```python
# insights

#1. team wise driver count
team_insights_df = df.groupBy("team_name") \
    .count() \
    .withColumnRenamed("count", "driver_count")

display(team_insights_df)
```

```python
# to save
team_insights_df.write \
    .format("delta") \
    .mode("overwrite") \
    .save("abfss://presentation@driverf1storageprject.dfs.core.windows.net/team_driver_count")
```

```python
#1. “Top Drivers by Overall Session Participation”

from pyspark.sql.functions import count

top_drivers = df.groupBy("full_name", "team_name", "name_acronym") \
    .agg(
        count("session_key").alias("total_sessions")
    ) \
    .orderBy("total_sessions", ascending=False)

display(top_drivers)
```

```python
top_drivers.write.format("delta") \
    .mode("overwrite") \
    .save(
        "abfss://presentation@driverf1storageprject.dfs.core.windows.net/top_drivers"
    )
```

```python
# to convert csv format

top_drivers.coalesce(1).write \
    .mode("overwrite") \
    .option("header", "true") \
    .csv(
        "abfss://presentation@driverf1storageprject.dfs.core.windows.net/csv/top_drivers"
    )
```

```python
# 2.“Driver Participation Across Sessions”

from pyspark.sql.functions import count

driver_perf = df.groupBy("full_name", "name_acronym", "team_name") \
    .agg(
        count("session_key").alias("total_sessions")
    ) \
    .orderBy("total_sessions", ascending=False)

display(driver_perf)
```

```python
# to save into delta form

driver_perf.write.format("delta") \
    .mode("overwrite") \
    .save(
        "abfss://presentation@driverf1storageprject.dfs.core.windows.net/driver_participation"
    )
```

```python
# to save into csv form

driver_perf.coalesce(1).write \
    .mode("overwrite") \
    .option("header", "true") \
    .csv(
        "abfss://presentation@driverf1storageprject.dfs.core.windows.net/csv/driver_participation"
    )
```

```python
#3. “Team Performance Based on Driver Participation”

from pyspark.sql.functions import count

team_perf = df.groupBy("team_name", "team_colour") \
    .agg(
        count("driver_number").alias("total_drivers"),
        count("session_key").alias("total_sessions")
    ) \
    .orderBy("total_sessions", ascending=False)

display(team_perf)
```

```python
team_perf.write.format("delta") \
    .mode("overwrite") \
    .save(
        "abfss://presentation@driverf1storageprject.dfs.core.windows.net/team_participation"
    )
```

```python
# to save intom csv
team_perf.coalesce(1).write \
    .mode("overwrite") \
    .option("header", "true") \
    .csv(
        "abfss://presentation@driverf1storageprject.dfs.core.windows.net/csv/team_participation"
    )
```

```python
#4. “Driver Distribution Across Teams”
driver_team = df.select("full_name", "name_acronym", "team_name", "team_colour") \
    .dropDuplicates() \
    .orderBy("team_name")

display(driver_team)
```

```python
driver_team.write.format("delta") \
    .mode("overwrite") \
    .save(
        "abfss://presentation@driverf1storageprject.dfs.core.windows.net/driver_team_distribution"
    )
```

```python
driver_team.coalesce(1).write \
    .mode("overwrite") \
    .option("header", "true") \
    .csv(
        "abfss://presentation@driverf1storageprject.dfs.core.windows.net/csv/driver_team_distribution"
    )
```

```python
# 5.team vise driver count
from pyspark.sql.functions import countDistinct

team_driver_count = df.groupBy("team_name") \
    .agg(
        countDistinct("driver_number").alias("driver_count")
    ) \
    .orderBy("driver_count", ascending=False)

display(team_driver_count)
```

```python
team_driver_count.write.format("delta") \
    .mode("overwrite") \
    .save(
        "abfss://presentation@driverf1storageprject.dfs.core.windows.net/team_driver_count"
    )
```

```python
team_driver_count.coalesce(1).write \
    .mode("overwrite") \
    .option("header", "true") \
    .csv(
        "abfss://presentation@driverf1storageprject.dfs.core.windows.net/csv/team_driver_count"
    )
```
